```python
from scipy import stats
```

```python
# 1. Thực hiện kiểm định trung bình của 1 biến số (định lượng) bằng phương pháp
# Z-Test bằng một giá trị nào đó với mức sai lầm là 5%. Cho nhận xét.
from statsmodels.stats.weightstats import ztest as ztest
```

```python
# Kiểm tra xem trung bình Engine Cylinders có bằng 6.0 với alpha = 0.05
ztest(df['Engine Cylinders'], value=6.0)
# Kết luận: trung bình không bằng 6.0 VÌ P-VALUE=1.0823607284220433e-113 < alpha=0.05
```

```python
# 2. Thực hiện kiểm định trung bình của 2 biến số (định lượng) có bằng nhau không
# bằng phương pháp T-test với mức sai lầm là 10%. Cho nhận xét   


# Kiểm tra xem trung bình Engine Cylinders của Honda và trung bình Number of Doors có bằng nhau hay không với alpha = 0.1
# H0: mean Engine Cylinders = mean Number of Doors = 0

dfH = df[df['Make'] == "Honda"]

stats.ttest_ind(dfH['Engine Cylinders'], dfH['Number of Doors'], equal_var=True)

# Kết luận: Do alpha = 0.1 > p-value = 2.8686661056466756e-63
# => Từ chối H0 ở mức ý nghĩa 10%
# => Suy ra không có đủ dữ liệu để nói rằng trung bình Engine Cylinders bằng trung bình Number of Doors hay nói cách khác là từ chối H0 ở mức sai lầm 10%
```

```python
# 3. Thực hiện kiểm tra 2 biến định lượng có tương quan với nhau không bằng 
# phương pháp Chi-Square với mức sai lầm là 5%? Cho nhận xét.

# Thực hiện kiểm tra 2 biến định lượng có tương quan với nhau không bằng
# phương pháp Chi-Square với mức sai lầm là 5%? Cho nhận xét.   


# H0: city mpg và number of doors không có quan hệ với nhau
from scipy.stats import chi2_contingency

crosstab = pd.crosstab(df['city mpg'], df['Number of Doors'], rownames=['city mpg'], colnames=['Number of Doors'])
crosstab

# Kiểm định trung bình cho biến Number of Doors và city mpg xem 2 biến có tương quan với nhau ko
```

```python
stat, p, dof, expected = chi2_contingency(crosstab)

# interpret p-value
alpha = 0.05
print("p value is " + str(p))

if p <= alpha:
  print("Dependent (Reject H0) => có quan hệ với nhau")
else:
  print("Independent (H0 holds true) => Độc lập với nhau") 
```
